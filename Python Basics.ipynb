#### Dictionaries: Datatype in python, unordered, Mutable and Indexed declared by {} ideally ky value pairs

```python
dic = {"Employee1" : "Phanindra", "Employee2" : "Gopalrao"}
```

```python
dic
```

```python
type(dic)
```

```python
# Indexing can be done on KEY or Value pair
dic["Employee2"]
```

```python
dic2 = {"Car1": "Suzuki", "Car2":"Toyota","Car3":"Hundai","Car4":"Ford"}
```

```python
# Traversing over KEYS
for x in dic2:
    print(dic2[x])
```

```python
# Traversing through VALUES
for x in dic2.values():
    print(x)
```

```python
for x in dic2.items():
    print(x)
```

```python
# UPdate new key and value
dic2["Car4"] = "Ford"
```

```python
dic2["Car5"] = "Jeep"
```

```python
dic2
```
